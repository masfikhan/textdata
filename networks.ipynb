## Network Tools

```python
# initialize graph
disney = nx.Graph()

# add edges
disney.add_edge('Selena Gomez','Demi Lovato')
disney.add_edge('Demi Lovato','Miley Cyrus')
disney.add_edge('Sprouse Twins','Selena Gomez')
disney.add_edge('Sprouse Twins','Miley Cyrus')
disney.add_edge('Demi Lovato','Joe Jonas')
```

```python
# draw graph 
nx.draw_networkx(disney)
```

```python
# remove edge
disney.remove_edge('Demi Lovato', 'Joe Jonas')
nx.draw_networkx(disney)
```

```python
# degree of node
nx.degree(disney, 'Selena Gomez')
```

```python
# weighted network

disney_weighted = nx.Graph()

disney_weighted.add_edge('Selena Gomez','Demi Lovato', weight=5)
disney_weighted.add_edge('Demi Lovato','Miley Cyrus', weight=1)
disney_weighted.add_edge('Sprouse Twins','Selena Gomez', weight=2)
disney_weighted.add_edge('Sprouse Twins','Miley Cyrus', weight=10)

nx.draw_networkx(disney_weighted)
```
